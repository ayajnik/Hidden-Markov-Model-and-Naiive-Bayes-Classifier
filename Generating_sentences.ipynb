```python
def read():
    global initial
    global jo
    initial = input("Enter the name of the text file : ")
    jo = open(initial, encoding='utf-8').read()
    return jo
```

```python
read()
```

# **Over here, we split each character in our text file into a single component and convert it into a list.**

```python
##split the data into words
corpus = jo.split()
print(corpus)
def make_pairs(corpus):
    for i in range(len(corpus) - 1):
        yield(corpus[i], corpus[i+1])

pairs = make_pairs(corpus)
```

# **The pairs that we made in the previous code snippet, we now make pairs of them using a dictionary. This is necessary in order to make our probabilistic model.**

```python
##appending the dictionary

word_dict = {}

for word_1, word_2 in pairs:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

# **Building our Hidden Markov Model.**

```python
##building the markov model

first_word = np.random.choice(corpus)

while first_word.islower():
    first_word = np.random.choice(corpus)

chain = [first_word]

n_words = 20

for i in range(n_words):
    chain.append(np.random.choice(word_dict[chain[-1]]))
```

# **Over here we make our predictions. We store our predictions in an object called (xy) and in the end, we print it.**

```python
##prediction

xy = (' '.join(chain))
print(xy)
```
